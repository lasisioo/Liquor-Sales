# Getting started

Once you've chosen your scenario, download the data from [the Iowa website](https://data.iowa.gov/Economy/Iowa-Liquor-Sales/m3tr-qhgy) in csv format. Start by loading the data with pandas. You may need to parse the date columns appropriately.

<b>Problem Statement:</b> 

Iowa State Governemnt provides information on sales of liquor stores across counties in the state. Asidd from using the data to make current analysis, can we also use it to make predictions?

```python
import pandas as pd
import numpy as np
import statsmodels.formula.api as smf
from sklearn import linear_model
from sklearn.metrics import r2_score
```

```python
df = pd.read_csv("../assets/Iowa_Liquor_sales_sample_10pct.csv")
df.head()
```

First, I count the number of rows in each column to see if they are equal to one another.

```python
df.count()
```

From the count above, we see that the numbers of observations in columns are not equal (this means that there are some missing entries in this data frame). Because this is a pretty large dataset, I can drop rows with missing variables without greatly affecting the data. Dropping the rowns with missing data makes all the columns equal in length. I also drop variable that will not be necessary for my analyses.

```python
#First, I drop the rows with the missing entries, and then the "Category" and the "Item number" columns
df2 = df.dropna()
df2.drop(["Category", "Item Number", "Vendor Number", "Volume Sold (Gallons)"], axis=1, inplace=True)
df2.count()
```

Now I want to look at what types of data are in each column.

```python
df.dtypes
```

Let transform the data to datetime and change the Cost, Retail and Sale values from an object to a float

```python
df2["Date"] = pd.to_datetime(df2["Date"], infer_datetime_format=True)

def getType(field):
    return field.replace("$", "")

df2["Sale (Dollars)"] = df2["Sale (Dollars)"].apply(getType).astype(float)
df2["State Bottle Cost"] = df2["State Bottle Cost"].apply(getType).astype(float)
df2["State Bottle Retail"] = df2["State Bottle Retail"].apply(getType).astype(float)
```

Further caluculations gives us a little bit more information about out data

```python
df2["Price/Liter"] = df2["State Bottle Retail"] / (df2["Bottle Volume (ml)"]/1000)
df2["Profit (Dollars)"] = df2["State Bottle Retail"] - df2["State Bottle Cost"]
```

```python
#Lets look at the dataframe to make sure it looks exactly how we want it to:
df2.head()
```

# Mine the data
Now you are ready to compute the variables you will use for your regression from the data. For example, you may want to
compute total sales per store from Jan to March of 2015, mean price per bottle, etc. Refer to the readme for more ideas appropriate to your scenario.

Pandas is your friend for this task. Take a look at the operations [here](http://pandas.pydata.org/pandas-docs/stable/groupby.html) for ideas on how to make the best use of pandas and feel free to search for blog and Stack Overflow posts to help you group data by certain variables and compute sums, means, etc. You may find it useful to create a new data frame to house this summary data.

To predict how stores will perform in 2016, I will first build a model using data from the first quarter of 2015 to make predictions for year end 2015, and then apply the same model to 2016 Q1 values to predict values for 2016 year end. I start off by slicing the data into 3: 2015, 2015Q1, and 2016Q1.

```python
#2015
df2.sort_values(by = ["Store Number", "Date"], inplace = True)
start = pd.Timestamp("20150101")
end = pd.Timestamp("20151231")
mask = (df2["Date"] >= start) & (df2["Date"] <= end)
sales2015 = df2[mask]
sales2015 = sales2015.groupby(by = ["Store Number"], as_index = False)
sales2015 = sales2015.agg({"Sale (Dollars)": [np.sum, np.mean],
                   "Volume Sold (Liters)": [np.sum, np.mean],
                   "Profit (Dollars)": np.mean,
                   "Price/Liter": np.mean,
                   "Zip Code": lambda x: x.iloc[0],
                   "City": lambda x: x.iloc[0],
                   "County Number": lambda x: x.iloc[0]})
sales2015.columns = [' '.join(col).strip() for col in sales2015.columns.values]
sales2015.columns = ["Store_Number", "City", "2015_Sales_Sum", "2015-Sales_Mean", "County_Number", 
                     "Zip_Code", "2015_Profit_Mean", "2015_Volume Sold_liters_Sum", "2015_Price_per_Liter_Mean", 
                     "2015_Volume_Sold_liters_Mean"]

sales2015.head()
```

```python
#2015 Q1
df2.sort_values(by = ["Store Number", "Date"], inplace = True)
start = pd.Timestamp("20150101")
end = pd.Timestamp("20150331")
mask = (df2["Date"] >= start) & (df2["Date"] <= end)
salesQ12015 = df2[mask]
salesQ12015 = salesQ12015.groupby(by = ["Store Number"], as_index = False)
salesQ12015 = salesQ12015.agg({"Sale (Dollars)": [np.sum, np.mean],
                   "Volume Sold (Liters)": [np.sum, np.mean],
                   "Profit (Dollars)": np.mean,
                   "Price/Liter": np.mean,
                   "Zip Code": lambda x: x.iloc[0],
                   "City": lambda x: x.iloc[0],
                   "County Number": lambda x: x.iloc[0]})
salesQ12015.columns = [' '.join(col).strip() for col in salesQ12015.columns.values]
salesQ12015.columns = ["Store_Number", "City", "2015Q1_Sales_Sum", "2015Q1_Sales_Mean", "County_Number", 
                     "Zip_Code", "2015Q1_Profit_Mean", "2015Q1_Volume Sold_liters_Sum", "2015Q1_Price_per_Liter_Mean", 
                     "2015Q1_Volume_Sold_liters_Mean"]
salesQ12015.head()
```

I merge the 2015 and 2015Q1 datasets on Store Number, City, and County Number, and then select the models I want to keep.

```python
merged2015 = pd.merge(salesQ12015, sales2015, on=["Store_Number", "City", "County_Number"])
merged2015.head()
```

```python
model2015 = merged2015[["Store_Number", "City", "2015Q1_Sales_Sum", "2015_Volume_Sold_liters_Mean", 
                        "2015Q1_Price_per_Liter_Mean", "2015_Sales_Sum", "Zip_Code_y", "2015_Profit_Mean", 
                        "2015_Price_per_Liter_Mean"]]
```

```python
#2016 Q1
df2.sort_values(by = ["Store Number", "Date"], inplace = True)
start = pd.Timestamp("20160101")
end = pd.Timestamp("20160331")
mask = (df2["Date"] >= start) & (df2["Date"] <= end)
sales2016 = df2[mask]
sales2016 = sales2016.groupby(by = ["Store Number"], as_index = False)
sales2016 = sales2016.agg({"Sale (Dollars)": [np.sum, np.mean],
                   "Volume Sold (Liters)": [np.sum, np.mean],
                   "Profit (Dollars)": np.mean,
                   "Price/Liter": np.mean,
                   "Zip Code": lambda x: x.iloc[0],
                   "City": lambda x: x.iloc[0],
                   "County Number": lambda x: x.iloc[0]})
sales2016.columns = [' '.join(col).strip() for col in sales2016.columns.values]
sales2016.columns = ["Store_Number", "City", "2016_Sales_Sum", "2016_Sales_Mean", "County_Number", 
                     "Zip_Code", "2016_Profit_Mean", "2016_Volume_Sold_liters_Sum", "2016_Price_per_Liter_Mean", 
                     "2016_Volume_Sold_liters_Mean"]


sales2016.head()
```

# Build your models

Using scikit-learn or statsmodels, build the necessary models for your scenario. Evaluate model fit.

I use linear regression to run the data in my merged table. This builds the model and then runs the 2016 values to give a predicted values for year end 2015.

```python
X = model2015[["2015Q1_Sales_Sum", "2015Q1_Price_per_Liter_Mean", "2015_Volume_Sold_liters_Mean"]]
y = model2015["2015_Sales_Sum"]

X2016 = sales2016[["2016_Sales_Sum", "2016_Price_per_Liter_Mean", "2016_Volume_Sold_liters_Mean"]]

lm = linear_model.LinearRegression()
model = lm.fit(X, y)
predictions = lm.predict(X)
predictions2016 = lm.predict(X2016)

sales2016["2016_Predicted_Sales"] = predictions2016

plt.scatter(y, predictions)
plt.xlabel("True Values")
plt.ylabel("Predictions")

print "Score:", model.score(X, y)
```

The plot shows a strong postive correlation between the true values and predicted values, with an rSquare value of 94%.

```python
print sales2016["2016_Predicted_Sales"].sum()
print sales2015["2015_Sales_Sum"].sum()
```

```python
print sales2016["2016_Predicted_Sales"].shape
print sales2015["2015_Sales_Sum"].shape
```

From the results, we see that predicted sales for 2016 are slightly lower than that of 2015. We should however keep in mind that the number of entries for each year are not the same.

Now we run the data in 2015 using cross validation and apply the model to 2016.

```python
from sklearn.cross_validation import train_test_split
from sklearn.cross_validation import cross_val_score, cross_val_predict
from sklearn import metrics
import statsmodels.formula.api as smf
```

```python
model_2015 =  model2015[["2015Q1_Sales_Sum", "2015Q1_Price_per_Liter_Mean", "2015_Volume_Sold_liters_Mean"]]
y = model2015["2015_Sales_Sum"]

X_train, X_test, y_train, y_test = train_test_split(model_2015, y, test_size=0.75)
print X_train.shape, y_train.shape
print X_test.shape, y_test.shape

model_2016 = sales2016[["2016_Sales_Sum", "2016_Price_per_Liter_Mean", "2016_Volume_Sold_liters_Mean"]]

lm2 = linear_model.LinearRegression()
model2 = lm2.fit(X_train, y_train)
predictions2 = lm2.predict(X_test)

scores = cross_val_score(model2, model_2015, y, cv=12)
print "Cross Validated Scores:", scores

Predictions3 = cross_val_predict(model2, model_2015, y, cv=12)
Accuracy = metrics.r2_score(y, Predictions3)
print "Cross Predicted Accuracy:", Accuracy

plt.scatter(y_test, predictions2)
plt.xlabel("True Values")
plt.ylabel("Predictions")

print "Score:", model.score(X_test, y_test)     
```

The plot shows astrong positive correlation between the predictions and true values, with an rSquare value of almost 94%.
